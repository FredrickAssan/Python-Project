```python
print ("==== STUDENT PROFILE SYSTEM ===")

name = input("Enter your full name:")
age = int(input("Enter your age:"))
gpa = float(input("Enter your GPA:"))
courses = input("Enter your courses (comma-separated): ").split(",")


student_profile = {
    "name": name,
    "age" : age,
    "gpa" : gpa,
    "courses" : courses,
    "num_courses": len(courses)
}

is_deans_list = gpa >= 3.5
if age < 18:
    status = "Minor"
else:
    status = "Adult"

print ("\n=== STUDENTS DETAILS===")

print(f"Name:{student_profile['name']}")

print(f"Age:{student_profile['age']}({status})")

print(f"GPA:{student_profile['gpa']}")

print(f"Courses:{student_profile['courses']}")

print(f"Number of Courses:{student_profile['num_courses']}")

print(f"Dean's List:{'Yes' if is_deans_list else 'No'}")


print("\n===Data Types Used===")
print("name:", type(name))
print("age:", type(age))
print("gpa:", type(gpa))
print("courses:", type(courses))
print("student_profile:", type(student_profile))
```
