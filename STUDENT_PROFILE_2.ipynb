```python
# ===== ASSIGNMENT 1: STUDENT PROFILE SYSTEM =====

# Step 1: Collect user input
name = input("Enter your full name: ")

age_input = input("Enter your age: ")
age = int(age_input)

gpa_input = input("Enter your GPA (0.0 - 4.0): ")
gpa = float(gpa_input)

courses_input = input("Enter your courses (comma-separated): ")
courses = [course.strip() for course in courses_input.split(",")]

# Step 2: Store data in a dictionary
student_profile = {
 "name": name,
    "age": age,
    "gpa": gpa,
    "courses": courses,
    "num_courses": len(courses)
}

# Step 3: Display formatted output
deans_list = gpa >= 3.5
status = "Minor" if age < 18 else "Adult"

print("\n===== STUDENT PROFILE =====")
print(f"Name: {student_profile['name']}")
print(f"Age: {student_profile['age']} ({status})")
print(f"GPA: {student_profile['gpa']}")
print(f"Courses: {student_profile['courses']}")
print(f"# Courses: {student_profile['num_courses']}")
print(f"Dean's List: {'Yes' if deans_list else 'No'}")
# Step 4: Demonstrate data types
print("\n===== DATA TYPES =====")
print(f"Name Type: {type(name)}")
print(f"Age Type: {type(age)}")
print(f"GPA Type: {type(gpa)}")
print(f"Dean's List Type: {type(deans_list)}")
print(f"Courses Type: {type(courses)}")
print(f"Student Profile Type: {type(student_profile)}")

# Step 5: Demonstrate type conversion
print("\n===== TYPE CONVERSION =====")
print(f"Original age input type: {type(age_input)}")
print(f"Converted age type: {type(age)}")
print(f"Original GPA input type: {type(gpa_input)}")
print(f"Converted GPA type: {type(gpa)}")
```
